```python
import numpy as np
import tensorflow as tf
```

```python
def label_images(picture_type,pic_labels):
    if picture_type=="cardinal":
        pic_labels = np.append(pic_labels,[[1,0,0]],axis=0)
        print("Cardinal")
    elif picture_type=="indigo":
        pic_labels = np.append(pic_labels,[[0,1,0]],axis=0)
        print("Indigo")
    else:
        pic_labels = np.append(pic_labels,[[0,0,1]],axis=0)
        print("Invalid Picture")
    #print("labels shape:",pic_labels.shape)
    return pic_labels
    
```

```python
#Directories
train_image_directory ="/tmp/training/"
valid_image_directory = "/tmp/validation/"  #Validation images
test_image_directory = "/tmp/testing/"
save_directory ="/tmp/Tensorflow_models/bird_classifier1.ckpt"
```

```python
#Important operations
init = tf.global_variables_initializer()
```

```python
#Features
Yfeatures =3  #Number of image classes
#Reshape
H=374 #Height
W=500 #Width
C =3 #Colour Channels
```

```python
train_img_file_names=tf.gfile.ListDirectory(train_image_directory)
valid_img_file_names=tf.gfile.ListDirectory(valid_image_directory)
test_img_file_names=tf.gfile.ListDirectory(test_image_directory)

#print("Training Images:",train_img_file_names)
#print("Validation Images:",valid_img_file_names)
#print("Testing Images:",test_img_file_names)

train_image_batch = np.zeros((1,H,W,C), dtype=np.uint8)  #Creating train image batch
train_pic_labels = np.zeros((1,Yfeatures), dtype=np.uint8)       #Creating label batch
valid_image_batch = np.zeros((1,H,W,C), dtype=np.uint8)  #Creating test image batch
valid_pic_labels = np.zeros((1,Yfeatures), dtype=np.uint8)      #Creating label batch
test_image_batch = np.zeros((1,H,W,C), dtype=np.uint8)  #Creating test image batch
test_pic_labels = np.zeros((1,Yfeatures), dtype=np.uint8)      #Creating label batch

#print(train_pic_label_batch)
with tf.Session() as sess:
    sess.run(init)
    #Store Training images as a numpy array
    for file in train_img_file_names:   #Iterate through files in directory and store
        print("Training image file name:",file)
        temp = file.split('.')
        temp= temp[0].split('_')
        picture_type = temp[0].lower()
        #picture_id = float(temp[1])
        train_pic_labels=label_images(picture_type,train_pic_labels)
            
        img_file_name= tf.gfile.FastGFile(train_image_directory+file)
        img_file = img_file_name.read()
        print("Training Image Original Shape:",sess.run(tf.shape(tf.image.decode_image(img_file))))#[Height, Width, Channels]
        image1 = sess.run(tf.reshape(tf.image.resize_image_with_crop_or_pad(tf.image.decode_image(img_file), H, W),shape=[1,H, W,C]))
        #print("Image New Shape:",image1.shape)
        train_image_batch = np.append(train_image_batch,np.divide(image1, 255),axis=0)# normalize and append to batch
    
    #Store Validation images as a numpy array
    for file in valid_img_file_names:   #Iterate through files in directory and store
        print("Validation image file name:",file)
        temp = file.split('.')
        temp= temp[0].split('_')
        picture_type = temp[0].lower()
        #picture_id = float(temp[1])
        valid_pic_labels=label_images(picture_type,valid_pic_labels)
            
        img_file_name= tf.gfile.FastGFile(valid_image_directory+file)
        img_file = img_file_name.read()
        print("Validation Image Original Shape:",sess.run(tf.shape(tf.image.decode_image(img_file))))#[Height, Width, Channels]
        image1 = sess.run(tf.reshape(tf.image.resize_image_with_crop_or_pad(tf.image.decode_image(img_file), H, W),shape=[1,H, W,C]))
        #print("Image New Shape:",image1.shape)
        valid_image_batch = np.append(valid_image_batch,np.divide(image1, 255),axis=0)# normalize and append to batch  
    
    #Store testing images as a numpy array
    for file in test_img_file_names:   #Iterate through files in directory and store
        print("Testing image file name:",file)
        temp = file.split('.')
        temp= temp[0].split('_')
        picture_type = temp[0].lower()
        #picture_id = float(temp[1])
        test_pic_labels=label_images(picture_type,test_pic_labels)
            
        img_file_name= tf.gfile.FastGFile(test_image_directory+file)
        img_file = img_file_name.read()
        print("Testing Image Original Shape:",sess.run(tf.shape(tf.image.decode_image(img_file))))#[Height, Width, Channels]
        image1 = sess.run(tf.reshape(tf.image.resize_image_with_crop_or_pad(tf.image.decode_image(img_file), H, W),shape=[1,H, W,C]))
        #print("Image New Shape:",image1.shape)
        test_image_batch = np.append(test_image_batch,np.divide(image1, 255),axis=0)# normalize and append to batch  


train_image_batch= np.delete(train_image_batch,0,0)
train_pic_labels= np.delete(train_pic_labels,0,0)
valid_image_batch= np.delete(valid_image_batch,0,0)
valid_pic_labels= np.delete(valid_pic_labels,0,0)
test_image_batch= np.delete(test_image_batch,0,0)
test_pic_labels= np.delete(test_pic_labels,0,0)
print("Train Image Batch Shape:",train_image_batch.shape)
print("Train Image Labels Shape:",train_pic_labels.shape)
print("Validation Image Batch Shape:",valid_image_batch.shape)
print("Validation Image Labels Shape:",valid_pic_labels.shape)
print("Testing Image Batch Shape:",test_image_batch.shape)
print("Testing Image Labels Shape:",test_pic_labels.shape)
#print(image_batch)
```
